#Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Age Group Data

```python
data = {
    'Age Group': ['13-17', '18-24', '25-34', '35+'],
    'Percentage': [30, 45, 20, 5],
}
```

```python
df = pd.DataFrame(data)
```

#Plot Age Group Distribution

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Age Group'], df['Percentage'], color=['#3498db', '#2ecc71', '#e74c3c', '#9b59b6'])
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

#Gender Data

```python
gender_data = {
    'Gender': ['Male', 'Female'],
    'Percentage': [60, 40],
}
```

```python
gender_df = pd.DataFrame(gender_data)
```

#Plot Gender Distribution

```python
plt.figure(figsize=(10, 6))
plt.pie(gender_df['Percentage'], labels=gender_df['Gender'], autopct='%1.1f%%', colors=['#3498db', '#e74c3c'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```
